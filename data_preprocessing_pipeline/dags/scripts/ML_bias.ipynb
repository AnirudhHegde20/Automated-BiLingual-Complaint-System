```python
# Bias threshold (example: if accuracy < 0.8 for any slice, trigger an alert)
bias_threshold = 0.8

# Calculate slice-specific metrics
slice_metrics = metric_frame.by_group

# Function to check for bias and trigger an alert if threshold is crossed
def check_for_bias(slice_metrics, threshold):
    alert_flag = False
    alerts = []
    
    for slice_name, accuracy in slice_metrics.items():
        if accuracy < threshold:
            alert_flag = True
            alerts.append(f"Bias Alert: Accuracy for slice '{slice_name}' is below threshold: {accuracy:.2f}")
    
    if alert_flag:
        print("\n--- BIAS ALERTS ---")
        for alert in alerts:
            print(alert)
    else:
        print("\nNo bias detected across slices.")


check_for_bias(slice_metrics, bias_threshold)
```
